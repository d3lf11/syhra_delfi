```python

def hitung_laju_alir(v, t):
  """
  Fungsi untuk menghitung laju alir minyak.

  Args:
    v: Volume minyak (dalam barrel).
    t: Waktu (dalam jam).

  Returns:
    Laju alir minyak (dalam barrel per jam).
  """

  laju_alir = v / t
  return laju_alir

# Masukkan nilai volume dan waktu
v = float(input("Masukkan volume minyak (dalam barrel): "))
t = float(input("Masukkan waktu (dalam jam): "))

# Hitung laju alir
laju_alir = hitung_laju_alir(v, t)

# Tampilkan hasil
print("Laju alir minyak:", laju_alir, "barrel per jam")

```

```python

def hitung_recovery_factor(No, N):
  """
  Fungsi untuk menghitung recovery factor (RF) menggunakan persamaan keseimbangan material.

  Args:
    No: Jumlah minyak yang diproduksi (dalam barrel).
    N: Jumlah minyak awal di tempat (dalam barrel).

  Returns:
    Recovery factor (RF).
  """

  m = 0.1  # Gas cap ratio

  RF = (No + m * N) / N
  return RF

# Masukkan nilai jumlah minyak yang diproduksi dan jumlah minyak awal
No = float(input("Masukkan jumlah minyak yang diproduksi (dalam barrel): "))
N = float(input("Masukkan jumlah minyak awal di tempat (dalam barrel): "))

# Hitung recovery factor
RF = hitung_recovery_factor(No, N)

# Tampilkan hasil
print("Recovery factor (RF):", RF)

```

```python

import numpy as np

def hitung_permeabilitas_efektif(Krw, Sw, Sor, n, m):
  """
  Fungsi untuk menghitung permeabilitas efektif menggunakan rumus Corey.

  Args:
    Krw: Permeabilitas relatif air.
    Sw: Saturasi air.
    Sor: Saturasi minyak residual.
    n: Eksponen Corey untuk air.
    m: Eksponen Corey untuk minyak.

  Returns:
    Permeabilitas efektif.
  """

  kr = Krw * (Sw ** n) * ((1 - Sor - Sw) ** m)
  return kr

# Masukkan nilai Krw, Sor, n, dan m
Krw = float(input("Masukkan permeabilitas relatif air (Krw): "))
Sor = float(input("Masukkan saturasi minyak residual (Sor): "))
n = float(input("Masukkan eksponen Corey untuk air (n): "))
m = float(input("Masukkan eksponen Corey untuk minyak (m): "))

# Buat array saturasi air (Sw) dalam range 0.1 hingga 0.9
Sw = np.arange(0.1, 0.9, 0.1)

# Hitung permeabilitas efektif untuk setiap nilai Sw
kr = hitung_permeabilitas_efektif(Krw, Sw, Sor, n, m)

# Tampilkan hasil
print("Saturasi air (Sw):", Sw)
print("Permeabilitas efektif (kr):", kr)

```

```python

import matplotlib.pyplot as plt
import numpy as np

# Data
jumlah_rig = [15, 18, 20, 22, 25, 27, 30, 32, 35, 37, 40, 42, 45, 47, 50, 52, 55, 57, 60, 65]
volume_ekspor = [320, 340, 360, 380, 400, 440, 480, 640, 700, 740, 800, 840, 900, 940, 1000, 1040, 1100, 1140, 1200, 1300]

# Membuat diagram scatter
plt.figure(figsize=(6, 6))  # Mengatur ukuran diagram
plt.scatter(jumlah_rig, volume_ekspor, s=50, c='blue', marker='o')  # Membuat scatter plot

# Menghitung garis regresi
slope, intercept = np.polyfit(jumlah_rig, volume_ekspor, 1)
x_fit = np.array([min(jumlah_rig), max(jumlah_rig)])
y_fit = slope * x_fit + intercept

# Menggambar garis regresi
plt.plot(x_fit, y_fit, color='black', linewidth=2)

# Memberikan label pada sumbu
plt.xlabel('Jumlah Rig Minyak Aktif')
plt.ylabel('Volume Ekspor Minyak')

# Memberikan judul diagram
plt.title('Hubungan Jumlah Rig Minyak Aktif dan Volume Ekspor Minyak')

# Menampilkan diagram
plt.grid(True)  # Menampilkan grid
plt.show()
```
